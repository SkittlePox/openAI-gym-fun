```python
%matplotlib notebook
import matplotlib.pyplot as plt
import gym
import random

env = gym.make('LunarLanderContinuous-v2')
print(env.action_space)
print(env.observation_space)
```

```python
plt.rcParams['animation.html'] = 'jshtml'
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
```

```python
x, y = [], []

observation = env.reset()
for t in range(100):
    env.render()
    action = random.randint(0, 3)
    observation, _, done, _ = env.step(action)
    
    x.append(t)
    y.append(observation[1])
    
    ax.plot(x, y, color='b')
    ax.plot(x, 1, color='r')
    fig.canvas.draw()
    ax.set_xlim(left=max(0, t-50), right=t+50)
    
    if done:
        break
env.close()
```

```python
observation = env.reset()
p_h = -1.5
p_v = 0.05

for t in range(500):
    env.render()
    
    x = observation[0]
    y = observation[1]
    
    h = x * p_h
    v = y * p_v
    
    action = [v, h]
    
    observation, reward, done, info = env.step(action)
    
    if done:
        break
env.close()
    
```
